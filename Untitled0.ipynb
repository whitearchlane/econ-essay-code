```python
import pandas as pd
import numpy as np

df = pd.read_csv('ncbirths.csv')

# Drop missing values in weeks and habit
df = df.dropna(subset=['weeks', 'habit'])

# Create premature variable
df['premature'] = (df['weeks'] < 37).astype(int)

# Contingency table
table = pd.crosstab(df['habit'], df['premature'])
print(table)

# Proportions per group
props = pd.crosstab(df['habit'], df['premature'], normalize='index')
print(props)
```

```python
smoker = df[df['habit'] == 'smoker']
nonsmoker = df[df['habit'] == 'nonsmoker']

n1 = len(smoker)
n2 = len(nonsmoker)
p1 = smoker['premature'].mean()
p2 = nonsmoker['premature'].mean()

diff = p1 - p2
se = np.sqrt((p1*(1-p1)/n1) + (p2*(1-p2)/n2))
z = 1.96
ci_lower = diff - z*se
ci_upper = diff + z*se

print(f'p1 smoker: {p1:.4f}')
print(f'p2 nonsmoker: {p2:.4f}')
print(f'Difference: {diff:.4f}')
print(f'95% CI: ({ci_lower:.4f}, {ci_upper:.4f})')

from statsmodels.stats.proportion import proportions_ztest
import numpy as np
count = np.array([smoker['premature'].sum(), nonsmoker['premature'].sum()])
nobs = np.array([n1, n2])
stat, pval = proportions_ztest(count, nobs)
print(f'Z statistic: {stat:.4f}')
print(f'P value: {pval:.4f}')
```
